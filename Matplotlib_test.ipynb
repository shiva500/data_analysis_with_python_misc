```python
%matplotlib inline
# %matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
fig=plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

# Save figures

```python
fig.savefig('my_figure.png')
```
