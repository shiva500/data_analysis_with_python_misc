```python
sales['Customer Sub Group']
#sales['Sale Order Quantity']

sales = sales.dropna()
```

```python
# get columns to plot
columns = sales.columns.drop(['Customer Sub Group'])
# create x data
x_data = range(0, sales.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, sales[column], label=column)
# set title and legend
ax.set_title('Sales Dataset')
ax.legend()
```

```python

# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(sales['Customer Sub Group'])
# set title and labels
ax.set_title('Sales Data')
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```
